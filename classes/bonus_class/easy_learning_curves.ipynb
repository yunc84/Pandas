Now we can import the package - we want to use the draw_learning_curves method

```python
from sk_modelcurves.learning_curve import draw_learning_curve
import matplotlib.pyplot as plt
%matplotlib inline
```

Let's make a fake model

```python
from sklearn.datasets import make_classification
from sklearn.datasets import make_regression
from sklearn.linear_model import Lasso
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_clf, y_clf = make_classification(n_samples=1000)
X_reg, y_reg = make_regression(n_samples=1000)
```

### Learning curve for regression with mean squared error:

```python
lasso = Lasso(alpha=10)
```

```python
draw_learning_curve(lasso, X_reg, y_reg, scoring='mean_squared_error');
```

### Learning curve for classification with precision using 5-fold CV

```python
knn = KNeighborsClassifier(n_neighbors=2)
```

```python
draw_learning_curve(knn, X_clf, y_clf, scoring='precision', cv=5);
```

### Multiple models! This might be useful for your HW2...

```python
knn5 = KNeighborsClassifier(n_neighbors=5)
knn50 = KNeighborsClassifier(n_neighbors=50)
```

```python
# must specify estimator_titles
draw_learning_curve([knn5, knn50], X_clf, y_clf, estimator_titles=['kNN with 5', 'kNN with 50'], cv=5,
                   scoring='accuracy')
# make the plot bigger
fig = plt.gcf()
fig.set_size_inches(8, 8)
```
