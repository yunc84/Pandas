Let's import our advertising data:

```python
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('data/sales.csv')
```

```python
df.head()
```

We can easily plot all the columns as lines:

```python
df.plot()
```

Let's make a line graph of just our marketing mix:

```python
df.drop('Sales', axis=1).plot(kind='line')
```

Let's check how TV trends with Sales:

```python
df.plot(kind='scatter', x='TV', y='Sales')
```

```python
df['Sales'].plot(kind='hist')
```

# Plotting with seaborn

First we need to install the seaborn package with `pip`

- Note: this is how we would install any Python package
- Note: this is not compatible with older versions of Windows

```python
!conda install seaborn
```

```python
import seaborn as sns
```

Seaborn allows us to quickly generate a line of best fit

```python
sns.lmplot(x='TV', y='Sales', data=df)
```

We can also generate the pairplot. Extremely powerful!

```python
sns.pairplot(df)
```
